```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from fairlearn.reductions import ExponentiatedGradient, DemographicParity, EqualizedOdds
from fairlearn.metrics import MetricFrame, selection_rate, false_positive_rate, false_negative_rate

# Load your COMPAS dataset
# Replace 'your_dataset.csv' with the actual file name and path
# compas_data = pd.read_csv('your_dataset.csv')

# # Assuming 'is_recid' is your target variable
# X = compas_data.drop('is_recid', axis=1)
# y = compas_data['is_recid']

# Step 1: Train-Test Split
# Split after oversampling
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.2, random_state=42)

# Step 2: Resampling - Oversampling
oversampler = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train, y_train)

# Alternatively, you can use undersampling
# undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)
# X_train_resampled, y_train_resampled = undersampler.fit_resample(X_train, y_train)

# Step 3: Train a Random Forest Classifier on the resampled data
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_resampled, y_train_resampled)

# Step 4: Evaluate the model on the test set
y_pred = rf_model.predict(X_test)
print("Classification Report on Test Set:")
print(classification_report(y_test, y_pred))

# Step 5: Fairness-aware Preprocessing (Re-weighting)
sensitive_features1 = X_train_resampled['race']  # Use resampled training set for sensitive features
estimator = RandomForestClassifier(random_state=42)

# Fairness constraints - Demographic Parity
constraint = DemographicParity()

# Train a fairness-aware model using Exponentiated Gradient Reduction
fairness_model = ExponentiatedGradient(estimator, constraint)
fairness_model.fit(X_train_resampled, y_train_resampled, sensitive_features=sensitive_features1)

sensitive_features2 = X_test['race']

# Evaluate fairness metrics on the original test set
predictions = fairness_model.predict(X_test)
fairness_metrics = MetricFrame(metrics={"selection_rate": selection_rate,"false_positive_rate": false_positive_rate,"false_negative_rate": false_negative_rate},
                               y_true=y_test,
                               y_pred=predictions,
                               sensitive_features=sensitive_features2)

print("\nFairness Metrics on Test Set:")
print(fairness_metrics.by_group)
```
